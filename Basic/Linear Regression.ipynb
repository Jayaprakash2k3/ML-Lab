```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

EggPuffs = {'PuffCount': [10,15,20,25,30,35,40,45,50], 'Cost': [150,220,250,390,490,500,570,640,800]}
df = pd.DataFrame(EggPuffs)
```

```python
plt.scatter(df["PuffCount"],df["Cost"])
plt.xlabel("PuffCount")
plt.ylabel("Cost")
plt.plot(df["PuffCount"],df["PuffCount"]*16, color='red', label='Rs.16')
plt.plot(df["PuffCount"],df["PuffCount"]*14, color='blue', label='Rs.14')
plt.plot(df["PuffCount"],df["PuffCount"]*15, color='green', label='Rs.15')
plt.legend()
plt.show()
```

```python
model = LinearRegression()
X= df[["PuffCount"]]
Y= df["Cost"]
model.fit(X,Y)
y_pred = model.predict(X)
plt.scatter(X, Y, label='Actual Data')
plt.plot(X, y_pred, 'r-', label='Linear Regression')
plt.xlabel('Puff Count')
plt.ylabel('Cost')
plt.title('Linear Regression for Egg Puffs')
plt.legend()
plt.show()
slope = model.coef_[0]
intercept = model.intercept_
print("Slope:", slope)
print("Intercept:", intercept)
```
