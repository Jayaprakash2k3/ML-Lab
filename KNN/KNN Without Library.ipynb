```python
from matplotlib import pyplot as plt
import pandas as pd
x=pd.read_csv('fake_bills.csv')
```

```python
import numpy as np
from collections import Counter

def euclidean_distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

def knn(train_data, test_data, train_labels, k):
    predicted_classes = []
    for test_point in test_data:
        distances = [euclidean_distance(test_point, train_point) for train_point in train_data]
        sorted_indices = np.argsort(distances)
        k_nearest_labels = [train_labels[i] for i in sorted_indices[:k]]
        most_common = Counter(k_nearest_labels).most_common(1)
        predicted_classes.append(most_common[0][0])
    
    return predicted_classes

# Example usage
train_data = x[["diagonal", "length"]].values
train_labels = x["is_genuine"].values
test_data = np.array([[171.36,113.51], [175.5,111.5]])
k = 3

predicted_classes = knn(train_data, test_data, train_labels, k)
print(predicted_classes)
```
