```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

data=pd.read_csv("data.csv")
kmeans = KMeans(n_clusters=3,n_init=10)
kmeans.fit(data[["Age","Income"]])
centers = kmeans.cluster_centers_
labels = kmeans.labels_
plt.scatter(data[["Age"]],data[["Income"]], c=labels, cmap='viridis', edgecolors='k')
plt.scatter(centers[:, 0], centers[:, 1], marker='X', s=100, linewidths=2, color='r', label='Cluster Centers')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('K-means Clustering')
plt.legend()
plt.show()
print(centers)
```
